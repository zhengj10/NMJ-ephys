```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_excel('/Users/zhengj10/Desktop/power.xlsx')
df = df.dropna()
df
```

```python
# df
plt.figure(figsize=[7,6])
plt.rc('font', size=13)
sns.set_style('whitegrid')
plt.plot(df.V.to_list(),df['660 nm'].to_list(),label='660 nm',color='r')
plt.plot(df.V.to_list(),df['530 nm'].to_list(),label='530 nm',color='g',linestyle='dashed')
plt.plot(df.V.to_list(),df['530 nm (ND6)'].to_list(),label='530 nm (ND6)',color='g')
plt.plot(df.V.to_list(),df['488 nm'].to_list(),label='488 nm',color='b')
plt.legend()
plt.xlabel('V',size=14)
plt.ylabel('\u03BCW',size=14)
plt.ylim([-2,50])
plt.tight_layout()
plt.savefig('/Users/zhengj10/Desktop/power.svg',format='svg')
plt.show()
```
